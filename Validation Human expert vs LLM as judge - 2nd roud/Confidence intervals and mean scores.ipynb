```python
results = []
t_star = 1.96
for (test, model_tech), group in filtered.groupby(['test', 'model_tech']):
    all_samples = group[['expert4', 'gemma3', 'llama', 'deepseek']].values.flatten()
    N = len(all_samples)
    
    mu_hat = np.mean(all_samples)
    sigma_hat = all_samples.std(ddof=1)
    
    lower = mu_hat - t_star * sigma_hat / np.sqrt(N)
    upper = mu_hat + t_star * sigma_hat / np.sqrt(N)
    
    results.append({
        'test': test,
        'model_tech': model_tech,
        'N': N,
        'mean': mu_hat,
        'ci_lower': lower,
        'ci_upper': upper,
        'ci': f"({lower:.3f}, {upper:.3f})"
    })

results_df = pd.DataFrame(results)
print(results_df)
```
